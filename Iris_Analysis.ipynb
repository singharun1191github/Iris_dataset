```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.drop(['Id'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Species'].str.replace('Iris-', '', regex=True)
```

```python
df.head()
```

```python
df['Species'] = df['Species'].astype(str).str.replace('Iris-', '', regex=False)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labencoder=LabelEncoder()
```

```python
df['Species_Encoder']=labencoder.fit_transform(df['Species'])
```

```python
df.head()
```

```python
df=df.drop(columns='Species', axis=1)
```

```python
df.head()
```

```python
df = df.rename(columns={'SepalLengthCm':'Sepal_Length','SepalWidthCm':'Sepal_Width','PetalLengthCm':'Petal_Length','PetalWidthCm':'Petal_Width'})
```

```python
df.isnull().sum()
```

```python
x = df.drop(columns='Species_Encoder')
y = df['Species_Encoder']
```

```python
x.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.20,random_state=42)
```

```python
x.shape, x_train.shape, x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(x_train,y_train)
```

```python
y_pred=logreg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,y_pred)
```

```python
accuracy
```
